```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OpenAIEmbeddings
```

```python

# Load existing ChromaDB
db = Chroma(
    persist_directory="../db/swim_rules_semantic",  # Path to your persisted ChromaDB
    embedding_function=OpenAIEmbeddings(model="text-embedding-3-small"),  # Ensure the same model used during creation
)
```

```python

# Retrieve all chunks with metadata
all_docs = db.get()  # Returns dict with 'ids', 'documents', 'metadatas', 'embeddings'

# Print results
for doc_id, text, metadata in zip(all_docs["ids"], all_docs["documents"], all_docs["metadatas"]):
    print(f"ID: {doc_id}\nText: {text[:100]}...\nMetadata: {metadata}\n---")
```
